Import necessary modules:


```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date, timedelta
```

Define list of companies

```python
companies = ['AAPL', 'GOOGL', 'AMZN', 'MSFT', 'META', 'TSLA', 'NVDA', 'JPM', 'JNJ', 'V']
```

Define time period for data retrieval


```python
today = date.today()
start_date = today - timedelta(days=2*365 + 1)
end_date = today
```

Data Pull

```python
# Create an empty DataFrame to store the stock price data
stock_price_data_10 = pd.DataFrame()

# Loop through the list of companies and retrieve stock price data
for company in companies:
    # Retrieve stock price data
    data = yf.download(company, start=start_date, end=end_date)
    
    # Add a column for company name or ticker symbol
    data['Company'] = company
    
    # Concatenate the retrieved data to the stock_price_data_10 DataFrame
    stock_price_data_10 = pd.concat([stock_price_data_10, data])

# Reset the index of the DataFrame
stock_price_data_10.reset_index(inplace=True)
```

```python
from google.colab import data_table

data_table.enable_dataframe_formatter()

# See Data
stock_price_data_10
```

```python
# Create a pivot table with Date as the index, Company as the columns, and Close as the values
close_prices = stock_price_data_10.pivot(index='Date', columns='Company', values='Close')

# Select the 'Date' column from the close_prices DataFrame
datelist = close_prices.index.to_list()

# Reset the index of the pivot table
close_prices.reset_index(inplace=True)

# Drop the first column
close_prices = close_prices.drop('Date', axis=1)

# Add the list of dates to a new 'Date' column in the close_prices DataFrame
close_prices['Date'] = datelist

# Set the 'Date' column as the new index of the close_prices DataFrame
close_prices = close_prices.set_index(['Date'])

# Convert the index of the close_prices DataFrame to a DateTimeIndex
close_prices.index = pd.to_datetime(close_prices.index)
```

```python
import plotly.graph_objects as go

# plotly
fig = go.Figure()

# set up ONE trace for each company
for col in close_prices.columns:
    fig.add_trace(go.Scatter(x=close_prices.index,
                             y=close_prices[col],
                             name=col,
                             visible=False))

# set the first trace to visible
fig.update_traces(visible=True)

# create the button to show all traces
buttons = [
    dict(label='All Companies',
         method='update',
         args=[{'visible': [True] * len(close_prices.columns)},
               {'title': 'Stock Prices for All Companies'}])
]

# create a button for each trace
for col in close_prices.columns:
    buttons.append(dict(label=col,
                        method='update',
                        args=[{'visible': [col == trace.name for trace in fig.data]},
                              {'title': f'Stock Prices for {col}'}]))

# add date slider
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label="1m", step="month", stepmode="backward"),
                dict(count=6, label="6m", step="month", stepmode="backward"),
                dict(count=1, label="YTD", step="year", stepmode="todate"),
                dict(count=1, label="1y", step="year", stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(visible=True),
        type="date"
    )
)

# some adjustments to the updatemenus
updatemenu = []
your_menu = dict()
updatemenu.append(your_menu)

# Update Menu Configuration
updatemenu[0]['buttons'] = buttons
updatemenu[0]['direction'] = 'down'
updatemenu[0]['showactive'] = True

# Set chart title and y-axis label
fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Closing Price (USD)'
)

# add dropdown menus to the figure
fig.update_layout(showlegend=True, updatemenus=updatemenu, title='Stock Prices for All Companies')

# Set the height and width of the plot window
fig.update_layout(height=700, width = 1400)

# Display the plot
fig.show()
```
